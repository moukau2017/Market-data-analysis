```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('MarketData.csv')
```

```python
for col in df.columns:
    print(col)
```

```python
df = df.drop(['Date'], axis=1)
```

```python
#dendogram plot
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
plt.figure(figsize=(8,4))
plt.title("Dendrograms")
dend = shc.dendrogram(shc.linkage(df, method='ward'))
```

```python
print(dend)
print(dend['color_list'])
```

In the above plot indicating that optimal number of clusters for the dataset is 2

```python
unique_color = set(dend['color_list'])
print(unique_color)
optimal_number_of_cluster=len(unique_color)-1
print(optimal_number_of_cluster)
```

```python
#Elbow method
error=[]
for k in range (1,11):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(df)
    error.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),error, 'o-')
plt.title("Elbow method")
plt.xlabel("No of clusters")
plt.ylabel("Error")
plt.show()
```

In the above plot the elbow is at k=2 indicating that optimal clusters for the dataset is 2
